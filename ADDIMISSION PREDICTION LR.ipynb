### Checking Outliers

```python
df
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    plt.boxplot(df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
# IQR
# Calculate the upper and lower limits
Q1 = df['TOEFL Score'].quantile(0.25)
Q3 = df['TOEFL Score'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
 
# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(df['TOEFL Score'] >= upper)[0]
lower_array = np.where(df['TOEFL Score'] <= lower)[0]
 
# Removing the outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
 
# Print the new shape of the DataFrame
print("New Shape: ", df.shape)
```

```python
# Calculate the upper and lower limits
Q1 = df['LOR'].quantile(0.25)
Q3 = df['LOR'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
 
# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(df['LOR'] >= upper)[0]
lower_array = np.where(df['LOR'] <= lower)[0]
 
# Removing the outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
 
# Print the new shape of the DataFrame
print("New Shape: ", df.shape)
```

```python
Q1 = df['Chance of Admit'].quantile(0.25)
Q3 = df['Chance of Admit'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
 
# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(df['Chance of Admit'] >= upper)[0]
lower_array = np.where(df['Chance of Admit'] <= lower)[0]
 
# Removing the outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
 
# Print the new shape of the DataFrame
print("New Shape: ", df.shape)
```

### Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
df.drop(['Serial No.'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x=df[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research']]
y=df['Chance of Admit']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
li=LinearRegression()
```

```python
li.fit(X_train,y_train)
```

```python
li.coef_
```

```python
cdf=pd.DataFrame(li.coef_,x.columns,columns=["coef"])
print(cdf)
```

```python
prediction=li.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import math
```

```python
print('Mean abosolute error:',mean_absolute_error(y_test,prediction))
print('Mean square error:',mean_squared_error(y_test,prediction))
print('RMSE:',math.sqrt(mean_squared_error(y_test,prediction)))
```

```python
residuals= y_test-prediction
```

```python
sns.distplot(residuals,bins=20)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,prediction)
```

```python
print(r2)  #accuracy of the model
```
